```python
from sklearn.datasets import make_blobs
```

## Загрузим выборки согласно 2 варианту

```python
X, y = make_blobs(n_samples=100, random_state=28, centers=7, cluster_std=2, shuffle=0)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

## Расстояние ближайшего соседа

```python
mergings_single = linkage(X, method='single')
mergings_single
```

```python
dendrogram(mergings_single)
plt.figure(figsize=(15, 8), dpi=200)
plt.show()
```

## Расстояние дальнего соседа

```python
mergings_complete = linkage(X, method='complete')
mergings_complete
```

```python
dendrogram(mergings_complete)
plt.figure(figsize=(15, 8), dpi=200)
plt.show()
```

## Расстояние Уорда

```python
mergings_ward = linkage(X, method='ward')
mergings_ward
```

```python
dendrogram(mergings_ward)
plt.figure(figsize=(15, 8), dpi=200)
plt.show()
```

## Выбор лучшего разбиения

```python
mergings_ward = linkage(X, method='ward')
mergings_ward
```

```python
dendrogram(mergings_ward)
plt.figure(figsize=(15, 8), dpi=200)
plt.show()
```

```python
import numpy as np

def update_cluster_centers(X, c):
    centers = np.zeros((4, 2))
    for i in range(1, 5):
        ix = np.where(c == i)
        centers[i - 1, :] = np.mean(X[ix, :], axis=1)
    return centers
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
T = fcluster(mergings_complete, 4, criterion='maxclust')
clusters = update_cluster_centers(X, T)
clusters
```

```python
plt.scatter(X[:, 0], X[:, 1], c=T)
plt.scatter(clusters[:, 0], clusters[:, 1], c='black')
```

## Вычисление характеристик

```python
from sklearn.metrics.pairwise import euclidean_distances
```

## Средняя сумма квадратов расстояний до центроида

```python
sum_sq_distance = np.zeros(4)
for i in range(1, 5):
    ix = np.where(T == i)[0]
    sum_sq_distance[i - 1] = np.sum(euclidean_distances(X[ix, :], [clusters[i - 1]]) ** 2)

sum_sq_distance = np.sum(sum_sq_distance) / 4
sum_sq_distance
```

## Средняя сумма средних внутрикластерных расстояний

```python
sum_avg_intercluster_distance = np.zeros(4)
for i in range(1, 5):
    ix = np.where(T == i)[0]
    sum_avg_intercluster_distance[i - 1] = np.sum(euclidean_distances(X[ix, :], [clusters[i - 1]]) ** 2) / len(X[ix, :])

sum_avg_intercluster_distance = np.sum(sum_avg_intercluster_distance) / 4
sum_avg_intercluster_distance
```

## Средняя сумма межкластерных расстояний

```python
sum_intercluster_distance = np.sum(euclidean_distances(clusters, clusters))
sum_intercluster_distance
```

## Кластеризация методом k-средних

```python
from sklearn.cluster import KMeans
```

```python
models = []
predicted = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init='auto')
    kmeans.fit(X)
    models.append(kmeans)
    predicted.append(kmeans.predict(X))
```

## Средняя сумма квадратов расстояний до центроида

```python
sum_avg_sq_distance = []
for it, kmean in enumerate(models):
    sum_avg_sq_distance.append(kmean.inertia_ / (it + 1))
sum_avg_sq_distance
```

```python
plt.plot(range(1, 11), sum_avg_sq_distance, '-o')
```

## Средняя сумма средних внутрикластерных расстояний

```python
new_centers = [kmean.cluster_centers_ for kmean in models]

sum_avg_intercluster_distance_avg = []
for k, kmean in enumerate(models):
    intercluster_sum = np.zeros(4)
    for i in range(4):
        ix = np.where(predicted[k] == i)[0]
        if len(ix) == 0:
            intercluster_sum[i - 1] = 0
        else:
            intercluster_sum[i - 1] = np.sum(euclidean_distances(X[ix, :], [kmean.cluster_centers_[i - 1]]) ** 2) / len(X[ix, :])
    sum_avg_intercluster_distance_avg.append(np.sum(intercluster_sum) / (k + 1))
sum_avg_intercluster_distance_avg
```

```python
plt.plot(range(1, 11), sum_avg_intercluster_distance_avg, '-o')
```

## Средняя сумма межкластерных расстояний

```python
sum_intercluster_distance_avg = []

for k, kmean in enumerate(models):
    value = np.sum(euclidean_distances(kmean.cluster_centers_, kmean.cluster_centers_))
    sum_intercluster_distance_avg.append(value / (k + 1))
sum_intercluster_distance_avg
```

```python
plt.plot(range(1, 11), sum_intercluster_distance_avg, '-o')
```
